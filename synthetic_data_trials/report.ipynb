As per description file provided lots of columns are converted into integer encoded categorical varibales
pandas consider those columns as type numerical..so changing the data types of those to categorical to perform chi square test

```python
# changing the dataTypes as per metadata
def identifyCategoricalColumsn(df,metadata):
    category_features = list(metadata.loc[metadata['DataType'] == 'object', 'Variable'])
    
    for each in category_features:
        df[each] = df[each].astype('object')
        
    return df
```

##### Using SDV:  The Synthetic Data Vault (SDV) is a Synthetic Data Generation ecosystem of libraries that allows users to easily learn single-table, multi-table and timeseries datasets to later on generate new Synthetic Data that has the same format and statistical properties as the original dataset.



#### Chi-Squared test
This metric uses the Chi-Squared test to compare the distributions
    of the two categorical columns. It returns the resulting p-value so that
    a small value indicates that we can reject the null hypothesis (i.e. and
    suggests that the distributions are different).
    
#### Inverted Kolmogorov-Smirnov D statistic
 the two-sample Kolmogorov–Smirnov test to compare
    the distributions of the two continuous columns using the empirical CDF.
    It returns 1 minus the KS Test D statistic, which indicates the maximum
    distance between the expteced CDF and the observed CDF values.
 As a result, the output value is 1.0 if the distributions are identical
    and 0.0 if they are completely different.

```python
# chi test based on categorical variables
def chi_KS_test(synthectic_data,real_data):
    return evaluate(synthectic_data, real_data, metrics=['CSTest','KSTest'], aggregate=False)
```

```python
#chi square test for dataset trial_data_A
chi_KS_test_A=chi_KS_test(trial_data_A, reference_data)
print(chi_KS_test_A)
```

```python
#chi square test for dataset trial_data_B
chi_KS_test_B=chi_KS_test(trial_data_B, reference_data)
print(chi_KS_test_B)
```

```python
#chi square test for dataset trial_data_C
chi_KS_test_C=chi_KS_test(trial_data_C, reference_data)
print(chi_KS_test_C)
```

#### trial_data_A has maximum score(P value) and KS test score..it is most similar to reference dataset according to chi square test and  KS test
